```python
# import libraries
import seaborn as sns # for data visualization
import matplotlib.pyplot as plt # for data visualization
import pandas as pd # for data analysis

# load dataset and create DataFrame ready to create heatmap
low = pd.read_csv("digit distribution of 12 different runs low.csv")
low = low.drop(columns=['Episode Code'], axis=1).set_index('Episode Name')

# set heatmap size
plt.figure(figsize= (16,9)) 
sns.set(font_scale=1.5)
# create heatmap seaborn

annot_kws={"fontsize":14
          }

cbar_kws = {"orientation":"vertical", 
            "shrink":1,
            'extend':'min', 
            'extendfrac':0.1, 
            "ticks":np.arange(0,201,20), 
            "drawedges":False,
           } # color bar keyword arguments

sns.heatmap(low, vmin = 0, vmax = 201, cmap="coolwarm", annot = True, linewidth = 2, cbar_kws=cbar_kws, annot_kws=annot_kws, fmt="d")

plt.title("RL agent - low set", fontsize = 50)
plt.xlabel("Digit", fontsize = 50)
plt.ylabel("Seed", fontsize = 50)
plt.rcParams["axes.labelsize"] = 30
plt.show()
```
